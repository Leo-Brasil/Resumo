```python
# Quando atribuimos uma váriavel e um operação a ela, temos um resultado boleano
idade = 19
idade >= 18
```

```python
# Vamos criar uma lista em branco para o preenchimento automatico dele e vamos criar uma lista para idades
# para cada idade percorrida irá fazer a pergunta, para cada idade percorrida na lista idades se maior que 18...

permissoes = []
idades = [10, 20, 30, 40, 50, 14, 17, 18, 16]

def permissao_para_dirigir (idades, permissoes):
  for idade in idades:
    if idade >= 18:
      permissoes.append(True)
    else:
      permissoes.append(False)

permissao_para_dirigir(idades, permissoes)
```

```python
permissoes
```

```python
# para cada permissao em permissoes:
#   se permissao == (verdadeiro):
#     imprimir (...)
#   se nao:
#     imprimir (...)

for permissao in permissoes:
  if permissao == (True):
    print('Tem permissão para dirigir')
  else:
    print('Não tem permissão para dirigir')
```
