# Tipo Booleano

```python
# Quando atribuimos uma váriavel e um operação a ela, temos um resultado boleano
idade = 19
idade >= 18
```

```python
idade = 14
idade >= 18
```

```python
# Vamos criar uma lista em branco para o preenchimento automatico dele e vamos criar uma lista para idades
# para cada idade percorrida irá fazer a pergunta, para cada idade percorrida na lista idades se maior que 18...

permissoes = []
idades = [10, 20, 30, 40, 50, 14, 17, 18, 16]

def permissao_para_dirigir (idades, permissoes):
  for idade in idades:
    if idade >= 18:
      permissoes.append(True)
    else:
      permissoes.append(False)

permissao_para_dirigir(idades, permissoes)
```

```python
permissoes
```

```python
# para cada permissao em permissoes:
#   se permissao == (verdadeiro):
#     imprimir (...)
#   se nao:
#     imprimir (...)

for permissao in permissoes:
  if permissao == (True):
    print('Tem permissão para dirigir')
  else:
    print('Não tem permissão para dirigir')
```

```python
# Plotamos as notas de matematica com a base 'x' como o numero de provas que o aluno fez e o eixo 'y' as notas dele

x = list(range(1,9))
y = notas_matematica
plt.plot(x,y)
```

```python
# Vamos fazer algumas configs nesse gráfico

x = list(range(1,9))                                    # Numero de provas
y = notas_matematica                                    # Notas de matematica
plt.plot(x,y, marker = 'o')                             # plotagem das notas e fazendo a bolinha do gráfico
plt.title('Notas de matemática')                        # Mudando o titulo
plt.xlabel('Provas')                                    # Mudando a legenda para provas
plt.ylabel('Notas')                                     # Mudando a legenda para notas
plt.show()                                              # Retirando a informação do armazenamento do gráfico
```

```python
notas_matematica = ['Matemática',8,7,6,6,7,7,8,10]
notas_portugues = ['Português',9,9,9,8,5,6,8,5]
notas_geografia = ['Geografia',10,10,6,7,7,7,8,7]

notas = [notas_matematica, notas_portugues, notas_geografia]
notas
```

```python
# Sabendo disso, qual alternativa abaixo irá gerar um gráfico para cada matéria?
for nota in notas:                 # Irei acessar cada lista 'Para cada nota na minha lista de notas'
  x = list(range(1,9))             # Faço um range da quantidade de provas
  y = nota[1:]                     # Com cada lista de nota acessada eu faço um slice para tirar as matérias da lista e pegar apenas as notas
  plt.plot(x, y, marker = 'o')     # Ploto no meu gráfico e faço a marcação com a bolinha
  plt.xlabel('Provas')             # Altero a legenda para provas
  plt.ylabel('Notas')              # Altero a legenda para notas
  plt.title(nota[0])               # Altero o titulo com a lista acessada eu pego apenas o primeiro item da lista que são as matérias 
  plt.show()                       # Retiro a legenda de armazenamento do meu gráfico
```
